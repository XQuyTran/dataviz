```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
survey_data = pd.read_excel('C://Users/Gateway/Desktop/energy_consuming.xlsx')
survey_data
```

```python
survey_data.shape
```

```python
survey_data.isna().sum()
```

```python
fig, ax = plt.subplots(1, 1, figsize=(15, 10))
ax = sns.heatmap(survey_data.corr(method='pearson'),annot=True)
```

```python
sns.lmplot(x='Number of occupants',y='Gas consumption',data=survey_data)
sns.lmplot(x='Number of occupants',y='energy per person',data=survey_data)
sns.lmplot(x='Number of occupants',y='total energy consumption',data=survey_data)
```

```python

sns.violinplot(x='House typology',y='total energy consumption',data=survey_data)
```

```python
sns.violinplot(x='Energy efficiency equipment',y='summer daylight',data=survey_data)
```

```python
sns.violinplot(x='Energy efficiency equipment',y='Winter daylight',data=survey_data)
```

```python
sns.swarmplot(x='Climatic region',y='number of airconditioning',data=survey_data)
```

```python
sns.swarmplot(x='Number of occupants',y='number of airconditioning',data=survey_data)
```

```python
sns.swarmplot(x='Number of floor',y='number of airconditioning',data=survey_data)
```

```python
melted = survey_data.melt(['House typology'],['House age','number of airconditioning'])
sns.swarmplot(x='House typology',y='value',data=melted,hue='variable',split=True)
plt.legend(bbox_to_anchor=(1,1),loc=2)
```

```python
melted = survey_data.melt(['Climatic region'],['House age','number of airconditioning'])
sns.swarmplot(x='Climatic region',y='value',data=melted,hue='variable',split=True)
plt.legend(bbox_to_anchor=(1,1),loc=2)
```

```python
melted = survey_data.melt(['Ownership'],['number of airconditioning','number of electricity water heater','Number of occupants'])
sns.swarmplot(x='Ownership',y='value',data=melted,hue='variable',split=True)
plt.legend(bbox_to_anchor=(1,1),loc=2)
```
